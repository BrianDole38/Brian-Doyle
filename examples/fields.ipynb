```python
import numba as nb
from parcels.numba.field import Particle, GridSpec, create_chunks, FieldSet, NumbaParticle, FieldUV, particle_inner_loop
from parcels.numba.record import create_record_pset
from parcels.numba.field import create_legacy_pset
from parcels import AdvectionRK4
import numpy as np
from matplotlib import pyplot as plt
from numba import njit
```

```python
np.random.seed(129837)
n_particles = 1000
```

```python
pset_legacy = create_legacy_pset(n_particles)
pset_record = create_record_pset(n_particles)
```

```python
grid = GridSpec((-5, 5, 1000, 10), (-5, 5, 1000, 10))
```

```python
def field_fun_v(lat_grid, lon_grid):
    alpha = np.arctan2(lon_grid, lat_grid)
    r = np.sqrt(lat_grid**2+lon_grid**2)
    v = -r*np.sin(alpha)
    return v

def field_fun_u(lat_grid, lon_grid):
    alpha = np.arctan2(lon_grid, lat_grid)
    r = np.sqrt(lat_grid**2+lon_grid**2)
    u = r*np.cos(alpha)
    return u
```

```python
fieldU = create_chunks(grid, pset_legacy, field_fun_u)
fieldV = create_chunks(grid, pset_legacy, field_fun_v)
```

```python
fieldset = FieldSet(grid, fieldU, fieldV)
```

```python
NumbaAdvection = njit(AdvectionRK4)
```

```python
%%time

particle_inner_loop(pset_legacy, fieldset, 100000)
```
